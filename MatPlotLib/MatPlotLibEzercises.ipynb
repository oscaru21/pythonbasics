Multiple line plot exercise
Your exercise is to create multiple line plots. You have the dataset 
which is the age and population related data regarding France and the US.

age = np.array([10, 20, 30, 40, 50, 60, 70, 80, 90])
population_france = np.array([100, 160, 310, 420, 490, 700, 675, 340, 80])
population_usa = np.array([140, 180, 350, 598, 780, 981, 1351, 780, 157])
Create line plots both for France as well as for the US on a single plot.

1) Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

2) Define Datasets

```python
age = np.array([10, 20, 30, 40, 50, 60, 70, 80, 90])
population_france = np.array([100, 160, 310, 420, 490, 700, 675, 340, 80])
population_usa = np.array([140, 180, 350, 598, 780, 981, 1351, 780, 157])
```

3)Define plot arguments

```python
usa_plot_properties = {'color':'red', 'fillstyle':'full'}
france_plot_properties = {'color':'blue', 'fillstyle':'full'}

plt.xlabel('Age')
plt.ylabel('Population')
plt.title('Population VS Age')

plt.plot(age, population_usa, **usa_plot_properties)
plt.fill_between(age, population_usa)

plt.plot(age, population_france, **france_plot_properties)
plt.fill_between(age, population_france)


plt.show()
```
