Scatter plots exercise:
In this exercise you have to visualize 2 different datasets on the same plot (with scatter plots). 
The dataset is given (Toyota and Hyundai related age and speed values):

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Toyota models age and top speed
x = np.array([1, 4, 6, 8, 10, 15, 20])
y = np.array([200, 180, 190, 150, 170, 130, 100])
 
# Hyundai models age and top speed
x2 = np.array([2, 4, 5, 9, 10, 13, 18])
y2 = np.array([190, 175, 132, 176, 112, 80, 95])
```

Let's visualize both of the datasets on a single plot. 
Use color green for Toyota and color orange for Hyundai. 
Let's use the legend function to show the brands.

```python
plt.scatter(x, y, color='green')

plt.scatter(x2, y2, color='orange')

plt.title('CARS AGE-TOP SPEED DIAGRAM')
plt.xlabel('Age')
plt.ylabel('Top Speed')
plt.legend(['Toyota', 'Hyundai'])

plt.show()
```
