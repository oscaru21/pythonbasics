MatPlotLib

```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

Create a basic NumPy data set:

```python
x = np.arange(10)
y = np.arange(5,15)
```

Plot the dataset:

```python
plt.plot(x, y)
plt.show()
```

Plotting a Sine function:

```python
#we have to crate an array with the values of x and y axis
#we can see a full cycle of the sine function from 0 to 2*pi
x_sin = np.arange(0, 2*math.pi, 0.1)
y_sin = np.sin(x_sin)

#now we plot the values
plt.plot(x_sin, y_sin)
plt.show()
```

Labels and Titles:

```python
title_font = dict(family='serif', color='red', size=17)
axis_font = dict(family='arial', color='green', size=17)

plt.xlabel('This is x', axis_font)
plt.ylabel('This is y', axis_font)
plt.title('f(x) = Sin(x)', title_font)

plt.plot(x_sin, y_sin)
plt.show()
```

Markers and lines:

```python
#we can define different marker style like '.' 'o' '>' '<' '*'
#and we can also define the size with ms argument
#we can find a list of available arguments here https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

#something that we could do is to create a configuration dictionary
plot_properties = {'marker': 'o', 'ms':3, 'mec':'green', 'mfc':'blue', 'color': 'red'}

#then we will pass each value using dictionary destructuring
plt.plot(x_sin, y_sin, **plot_properties)

#plt.plot(x_sin, y_sin, marker='o', ms=3)
plt.show()
```

Scatter Plots:

```python
#we can generate random values with the Numpy .random method

x = np.random.randint(100, size=1000)
y = np.random.randint(100, size=1000)
s = np.array([50 for num in range(1000)])

#instead of using the plot function we have to use the scatter
#we can assing color to the dots or use the cmap parameter to assign a range of colors predefines
#we need to define the color relationship, c=x+y or c=x or c=y

plt.scatter(x, y, sizes=s, c=x+y, cmap='inferno', alpha=0.5)
plt.colorbar()
plt.show()
```
